```python
#Fixer la valeur de gamma, alpha,matrices de récompences R
Gamma=0.8
alpha=1
R=np.array([[-1,-1,-1,-1,0,-1],[-1,-1,-1,0,-1,100],[-1,-1,-1,0,-1,-1],[-1,0,0,-1,0,-1],[0,-1,-1,0,-1,100],[-1,0,-1,-1,0,100]])
```

```python
#Initialisation de la matrice Q à zéro 
Q=np.zeros((6,6))
```

```python
num_ep=4000
final_state= R.shape[0]
```

```python
def QLearning(num_ep, Gamma, alpha, R, Q):
    
    for i in range (num_ep):
            final_state= Q.shape[0]
            state_initial=np.random.randint(0,final_state)
            actual_state=state_initial

            while (actual_state!=final_state-1):
                next_state = np.random.randint(0, final_state)
                while R[actual_state,next_state] <0:
                    next_state = np.random.randint(0, final_state)

                next_action=next_state
                Q[actual_state ,next_action]=Q[actual_state ,next_action]+alpha*(R[actual_state ,next_action]+(Gamma*max(Q[next_state]))-Q[actual_state,next_action])
                actual_state=next_state
    return Q

             
Gamma=0.9
alpha=1
R=np.array([[-1,-1,-1,-1,0,-1],[-1,-1,-1,0,-1,100],[-1,-1,-1,0,-1,-1],[-1,0,0,-1,0,-1],[0,-1,-1,0,-1,100],[-1,0,-1,-1,0,100]])
Q=np.zeros((6,6))
num_ep=4000              
Q=QLearning(num_ep, Gamma, alpha, R, Q)           
print(Q)
```

```python
ID101 = np.eye(27,27)*101
ID101 = ID101.astype(int)
ID2 = np.eye(27,27) # pour obtenir la matrice avec dernière colonne à 0 par produit matriciel
ID2[26,26] = 0
ID2 = ID2.astype(int)
T = np.zeros([27,27]) #pour obtenir la dernière colonne par pdt mat
T[26,26] = 1
T = T.astype(int)
ONE = np.ones([27,27])
ONE = ONE.astype(int)
# init R
R_Hanoi = - np.ones([27,27])
for i in range(0, 27, 1):
    a = i % 3;
    b = (i % 9) // 3;
    c = i // 9;
    # mise à jour des deux possibilités pour déplacer a (toujours possible car plus petit)
    aa = (a + 1) % 3
    aaa = (a + 2) % 3
    j = aa + 3*b + 9*c
    R_Hanoi[i,j] = 0
    j = aaa + 3*b + 9*c
    R_Hanoi[i,j] = 0
    # mise à jour du déplacement de b si possible
    if (a != b):
        bb = 3 - a - b
        j = a + 3*bb + 9*c
        R_Hanoi[i,j] = 0
    # mise à jour du déplacement de c si possbile
    if ((a == b) and (c !=a) ):
        cc = (2*c + a + b) % 3
        j = a + 3*b + 9*cc
        R_Hanoi[i,j] = 0
# mise à jour dernière colonne avec 100 quand déplacement possible
R_Hanoi = R_Hanoi.dot(ID2) + ((R_Hanoi.dot(T) + ONE).dot(T).dot(ID101) - ONE).dot(T) 
R_Hanoi[26,26] = 100
R_Hanoi = R_Hanoi.astype(int)
print(R_Hanoi)
```
